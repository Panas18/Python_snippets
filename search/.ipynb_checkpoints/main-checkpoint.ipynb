```python
import json
import sys
```

```python
with open("dataset/sentence_list.json") as f:
    sentence_list = json.load(f)
```

```python
def create_hash(sentence_list):
    resedue = [",", ".", "!", ":"]
    word_hash= {}
    for i, sentence  in enumerate(sentence_list):
        words = sentence.split(" ")
        words[-1] = words[-1].rstrip()
        for word in words:
            word = word.lower()
            if word[-1] in resedue:
                word = word[:-1]
            print(word)
            if word not in word_hash:
                word_hash[word] = [i]
            else:
                if not i in word_hash[word]:
                    word_hash[word].append(i)
    return word_hash
```

```python
word_hash = create_hash(sentence_list)
```
